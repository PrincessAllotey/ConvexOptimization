#1.1 Non-convex optimization - Gradient Ascent (20 points)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Initiate variables and constants:
D = [-5,5]
t = 100 
eta = [0.01, 0.02, 0.05]
x = np.zeros(t) #our x_vec
x_opt = (4*np.pi*(1/3)) + ((1/6)*np.arccos(-1/6))
print("Maximizer of f is x=" + str(x_opt))
```

###Starting from the point x_0 = 0, implement the gradient ascent algorithm with step sizes η ∈ {0.01, 0.02, 0.05} for 100 steps.

```python
#Function: 
def f_inv(y):
  return y + np.sin(6*y)

def f(x):
  return x + np.sin(6*x)

#Derivative of Function:
def df(x):
  return 1 + (6*np.cos(6*x))

#Gradient Ascent Algorithm:
def grad_ascent(eta_number,x_vec,t):
  for i in range(t-1):
    x_vec[i+1] = x_vec[i] + (df(x_vec[i])*eta_number)
  return x_vec 

f_opt = f(x_opt)
print("Maximum of f is: " + str(f_opt))
```

###Plot $f^{⋆} − f(x_t)$ versus the iteration number t where $f^{⋆}$ is the maximum of f on D for all step sizes on the same plot.

```python
# for j in eta:
#   plt.figure(figsize = (10,8))
#   plt.plot(f_opt - f(grad_ascent(j,x,t)))
#   plt.xlabel(r'iterations number (t)', fontsize = 15)
#   plt.ylabel(r'f^(⋆) − f(x_t)', fontsize = 15)
#   plt.title("f^(⋆) − f(x_t) versus t, (with eta = " + str(j) + ")")

plt.figure(figsize = (7,7))
plt.plot(f_opt - f(grad_ascent(eta[0],x,t)))
plt.plot(f_opt - f(grad_ascent(eta[1],x,t)))
plt.plot(f_opt - f(grad_ascent(eta[2],x,t)))
plt.legend(["eta = 0.01", "eta = 0.02", "eta = 0.05"])
plt.xlabel(r'iterations number (t)', fontsize = 15)
plt.ylabel(r'f^(⋆) − f(x_t)', fontsize = 15)
plt.title("f^(⋆) − f(x_t) versus t, (with eta = {0.01, 0.02, 0.05}")
```

###Plot $|x_t − x^{⋆}|$ (in log-scale) versus the iteration number t where $x^{⋆}$ is the maximizer of f on D for all step sizes on the same plot.

```python
plt.figure(figsize = (6,6))
plt.plot(np.log(np.absolute(x_opt - grad_ascent(eta[0],x,t))))
plt.plot(np.log(np.absolute(x_opt - grad_ascent(eta[1],x,t))))
plt.plot(np.log(np.absolute(x_opt - grad_ascent(eta[2],x,t))))
plt.legend(["eta = 0.01", "eta = 0.02", "eta = 0.05"])
plt.xlabel(r'iterations number (t)', fontsize = 15)
plt.ylabel(r'log|x^(⋆) − x_t|', fontsize = 15)
plt.title("log|x^(⋆) − x_t| versus t, (with eta = {0.01, 0.02, 0.05}")
```
